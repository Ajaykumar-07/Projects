Importing libraries

```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

Importing data set and extracting dependent and independent variable

```python
data_set = pd.read_csv("salary data simple linear regression.csv")

X = data_set.iloc[ 0:25 , 0:-1].values
y = data_set.iloc[ 0:25 , -1:]
```

Splitting into training set and test set 

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state = 42 )
```

Training on training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train,y_train)  # fits the dependent and independent variable and finds the best fit between both the variables
```

Predicting the test set

```python
y_predict = regressor.predict(X_test) #now this variable contains the predictions of the test data set
```

Visualizing the Training set

```python
plt.scatter(X_train,y_train, color = 'cyan')
plt.plot(X_train , regressor.predict(X_train) , color = 'red')
plt.title('CTC vs Experience')
plt.xlabel('experience')
plt.ylabel('CTC')
plt.show()
```

Visualizing the test data set

```python
plt.scatter(X_test , y_test , color = 'orange') #scatters the values of the test data set
#plt.scatter(X_test,y_predict,color = 'red')
plt.plot(X_train , regressor.predict(X_train) , color = 'blue')
plt.title('CTC vs Experience')
plt.xlabel('Experince')
plt.ylabel('CTC')
plt.show()
```
