Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing data set

```python
data = pd.read_csv('support vector regression.csv')
X = data.iloc[0:12 , 1:2].values
y = data.iloc[0:12 , 2].values
```

```python
print(y)
```

Feature scaling

```python
#since standardscalar class scales object of 2-D convert the y into a 2-D array first
y = y.reshape(len(y),1)

#transforming the y using standardscalar

from sklearn.preprocessing import StandardScaler
sc_y = StandardScaler()
sc_X = StandardScaler()

X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

training the Support vector regressioon model on the whole data set

```python
from sklearn.svm import SVR
regressor = SVR()

regressor.fit(X , y)
```

Predicting new results

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[6.5]])).reshape(-1,1))
```

Visualising the support vector regression results

```python
plt.scatter(sc_X.inverse_transform(X) , sc_y.inverse_transform(y) , color = 'cyan')
plt.plot(sc_X.inverse_transform(X) , sc_y.inverse_transform(regressor.predict(X).reshape(-1,1)) , color = 'green')
plt.title('Salary prediction using SVR')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()
```
