Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing data_set

```python
data_set = pd.read_csv('rbc data polynomial regression.csv')
X = data_set.iloc[ 0:26 , 0:1].values
y = data_set.iloc[0:26 , 1:2].values
```

Training the entire data on linear model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X , y)
```

Training the entire data on Polynomial Model

```python
#Adding polynomial features to the data set
from sklearn.preprocessing import PolynomialFeatures
X_poly = PolynomialFeatures(degree = 7).fit_transform(X)

#from sklearn.preprocessing import PolynomialFeatures
#poly_features = PolynomialFeatures(degree = 7)
#X_poly = poly_features.fit_transform(X)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly , y)
```

Visualising the linear model

```python
plt.scatter(X , y, color = 'red')
plt.plot(X , lin_reg.predict(X), color = 'orange')
plt.title('RBC vs AGE')
plt.xlabel('AGE')
plt.ylabel('RBC')
plt.show()
```

Visualising Polynomial model

```python
plt.scatter(X , y , color = 'red')
plt.plot(X , lin_reg2.predict(X_poly) , color = 'cyan')
plt.title('RBC vs AGE')
plt.xlabel('AGE')
plt.ylabel('RBC')
plt.show()
```

Prediction using linear model

```python
lin_reg.predict([[7]])
```

Prediction using polynomial model

```python
lin_reg2.predict(PolynomialFeatures(degree = 7).fit_transform([[7]]))
```
